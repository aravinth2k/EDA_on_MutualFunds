```python
# Import all required library for Data Extraction
import requests
import pandas as pd
from bs4 import BeautifulSoup
import json
```

#### Fetch Mutual fund urls

```python
# Fetch all mutual funds links from Groww
url_list=[]
for i in range(0,78):
    r=requests.get('https://groww.in/slr/v1/search/derived/scheme?available_for_investment=true&doc_type=scheme&max_aum=&page={}&plan_type=Direct&q=&size=15&sort_by=0'.format(str(i)))
    data=r.json()
    df2=pd.DataFrame(data['content'])
    df2['url']='https://groww.in/mutual-funds/'+df2.search_id
    url_list.extend(list(df2['url']))
```

```python
# Save all links in a text file
with open('links.txt','w',encoding='utf-8') as f:
    for link in url_list:
        f.write(link+'\n')
```

```python
#### Fetch Mutual Data Investment information
url_list = []
with open('links.txt','r',encoding='utf-8') as f:
    for link in f:
        url_list.append(link.strip())
```

```python
amc = []
l_date = []
rank = []
scheme_name = []
aum = []
invest_cat = []
fund_l_date = []
exit_load = []
expense_ratio = []
sub_cat = []
nav = []
min_inv_amount = []
min_sip_amount = []
simple_return = []
return1m = []
return1y = []
return2y = []
return3y = []
return4y = []
return5y = []
return10y = []

print('Data is Start Downloading ',end='')
for mf_link in url_list:
    try:
        r=requests.get(mf_link)
        print((url_list.index(mf_link)/len(url_list))*100)
        html_soup_object = BeautifulSoup(r.content, 'html.parser')
    
        json_data = html_soup_object.select('script#__NEXT_DATA__')[0].text
        json_format = json.loads(json_data)
        
        mf_info = json_format['props']['pageProps']['mf']['amc_info']
        amc.append(mf_info['amc'])
        l_date.append(mf_info['launch_date'])
        rank.append(mf_info['rank'])
        
        mf = json_format['props']['pageProps']['mf']
        scheme_name.append(mf['scheme_name'])
        aum.append(mf['aum'])
        invest_cat.append(mf['category'])
        fund_l_date.append(mf['launch_date'])
        exit_load.append(mf['exit_load'])
        expense_ratio.append(mf['expense_ratio'])
        sub_cat.append(mf['sub_category'])
        nav.append(mf['nav'])
        min_inv_amount.append(mf['min_investment_amount'])
        min_sip_amount.append(mf['min_sip_investment'])
        simple_return.append(mf['simple_return']['return_since_created'])
        return1m.append(mf['simple_return']['return1m'])
        return1y.append(mf['simple_return']['return1y'])
        return2y.append(mf['simple_return']['return2y'])
        return3y.append(mf['simple_return']['return3y'])
        return4y.append(mf['simple_return']['return4y'])
        return5y.append(mf['simple_return']['return5y'])
        return10y.append(mf['simple_return']['return10y'])
    except:
        continue
```

```python
df = pd.DataFrame({'AMC Name':amc, 'AMC Launch Date':l_date,'Rank':rank,'Mutual Fund Name':scheme_name,'AUM':aum,'Type':invest_cat,
'Scheme Launch Date':fund_l_date, 'Exit Load':exit_load,'Expense Ratio':expense_ratio,'Investment Type':sub_cat,'NAV':nav,
'Minimum Investment':min_inv_amount,'Minimum SIP Amount':min_sip_amount,'Total Return':simple_return,'1m_return':return1m,
'1y_return':return1y,'2y_return':return2y,'3y_return':return3y,'4y_return':return4y,'5y_return':return5y,'10y_return':return10y})
```

```python
df.to_csv(r'Data/MF_Data.csv',index=False)
```
